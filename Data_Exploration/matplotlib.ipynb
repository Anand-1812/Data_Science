```python
import matplotlib.pyplot as plt

x = [10, 30, 50, 70, 90]
y = [20, 40, 60, 80, 100]

plt.plot(x, y, color='blue', linewidth=2, marker='o', markersize=8, linestyle='--')
plt.title("Line Chart")
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")
plt.show();
```

```python
# Bar chart -> categorical data using rectangular boxes

a = ['Mon', 'Tue', 'Wed', 'Thurs', 'Fri']
b = [12, 15, 20, 25, 36]

plt.bar(a, b, color='blue', edgecolor='gray', linewidth=2)
plt.title("Bar Chart")
plt.ylabel("Day")
plt.xlabel("Values")
plt.show();
```

```python
# Histogram shows data by goruping them into bins, x-axis -> bin, y-axis -> frequency

x = [7, 8, 9, 10, 10, 12, 12, 12, 13, 14, 14, 15, 16, 16, 17, 18, 18, 19, 20, 20,
    21, 22, 23, 24, 25, 25, 26, 28, 30, 32, 35, 36, 38, 40, 42, 44, 48, 50]

plt.hist(x, bins=10, color='blue', edgecolor='gray', linestyle='--', alpha=0.5)
plt.title("Histogram")
plt.xlabel("Total Bill")
plt.ylabel("frequencies")
plt.show();
```

```python
# Scatter plaot -> relationship between variables

e = ['Mon', 'Tue', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun', 'Mon']
f = [170, 120, 250, 190, 180, 130, 260, 200]

size = [2, 3, 4, 2, 3, 2, 4, 3]  
bill = [170, 120, 250, 190, 180, 130, 260, 200] 

plt.scatter(e, f, c=size, s=bill, marker='D', alpha=0.5)
plt.title("Scatter plot")
plt.xlabel("Day")
plt.ylabel("Values")
plt.show();
```

```python
# Pie chart -> circular chart used to show data as %age

cars = ['AUDI', 'BMW', 'FORD','TESLA', 'JAGUAR',]
data = [23, 10, 35, 15, 12]
explode = [0.1, 0.5, 0, 0, 0.3]
colors = ['blue', 'green', 'orange', 'indigo', 'skyblue']

plt.pie(data, labels=cars, explode=explode, autopct='%1.2f', colors=colors, shadow=True)
plt.title("Pie Chart")
plt.show();
```

```python
# Box Plot -> Data is spread, it shows min, max, median, quartiles, also outliers

import matplotlib.pyplot as plt

data = [
    [10, 12, 14, 15, 18, 20, 22],
    [8, 9, 11, 13, 17, 19, 21],
    [14, 16, 18, 20, 23, 25, 27] 
  ]

plt.boxplot(data)
plt.xlabel("Groups")
plt.ylabel("Values")
plt.title("Box Plot")
plt.show()
```

```python
''' Heatmap -> values are shown as colors, helps to visualize pattern, correlations, intensity,
in a matrix like format '''

import numpy as npy

npy.random.seed(0) # ensure random value every time
data = npy.random.rand(10, 10) # 10*10 matrix of random numbers btw 0 and 1

plt.imshow(data, cmap='viridis', interpolation='nearest')

plt.colorbar()
plt.xlabel('X-axis')
plt.ylabel('Y-label')
plt.title('Heatmap')
plt.show()
```
