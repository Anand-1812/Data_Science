```python
import seaborn as sborn
import matplotlib.pyplot as plt

sborn.set_style('darkgrid')
custom_palette = sborn.color_palette("husl", 8)
sborn.set_palette(custom_palette)

data = sborn.load_dataset('iris')
sborn.pairplot(data, hue="species")
plt.show()
```

```python
# joint plot: combines scatter plot with distribution of the individual variables

data1 = sborn.load_dataset('tips')
sborn.jointplot(x='total_bill', y='tip', data=data1, kind='scatter', color='#007b7b')
plt.show()
```

```python
# Grid plot: for multiple subplots, how variable intercat with across different categories

plot = sborn.FacetGrid(data=sborn.load_dataset("tips"), col='time', row='sex')
plot.map(sborn.scatterplot, "total_bill", "tip")
plt.show()
```

```python
# visualizing linear relationship

# regplot -> plots a scatter plot along with a regression model fit

sborn.regplot(x='total_bill', y='tip', data=sborn.load_dataset("tips"), scatter_kws={'s':10}, line_kws={'color':'#1b1b2d'})
plt.show()
```
