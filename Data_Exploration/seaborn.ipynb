```python
# Built on top of matplotlib and pandas

import pandas as pd
import seaborn as sborn
import matplotlib.pyplot as plt

data = {'Name': ['ANSH', 'SAHIL', 'JAYAN', 'ANURAG'], 'Age': [21, 23, 20, 24]}
df = pd.DataFrame(data)

plt.figure(figsize=(6, 4))
sborn.lineplot(x='year', y='passengers', data=sborn.load_dataset('flights'), marker='o')
plt.title('Number of Passengers Over Time')
plt.show()
```

```python
# Scatter plot: relationship btw numerical values


data = {'Name': ['ANSH', 'SAHIL', 'JAYAN', 'ANURAG'], 'Age': [21, 23, 20, 45]}
df = pd.DataFrame(data)

# sborn.scatterplot(x=df.index, y='Age', data=df)
# plt.show()

# Box plot: qaurtiles vs cateogrical data

sborn.boxplot(y='Age', data=df)
plt.show()
```

```python
# Violin plot: similar as box plot but with more categorical data

# sborn.violinplot(y='Age', data=df)
# plt.show()

# swarm plot: display individual data points withotu overlap

sborn.swarmplot(x=df.index, y='Age', data=df, hue=1)
plt.show()
```

```python
# Bar plot: estimate of central tendency for a numerical value 

data = {'Name': ['ANSH', 'SAHIL', 'JAYAN', 'ANURAG'], 'Age': [21, 23, 20, 24]}
df = pd.DataFrame(data)

sborn.barplot(x='Name', y='Age', data=df)
plt.show()

# point plot: point estimates and confidence interval using scatter glyphs

sborn.pointplot(x='Name', y='Age', data=df)
plt.show()
```

```python
# count plot: disply the number of each category using bars to visualize

data = {'Name': ['ANSH', 'SAHIL', 'ANSH', 'JAYAN', 'ANURAG', 'ANURAG', 'ANURAG', 'SAHIL']}
df = pd.DataFrame(data)

sborn.countplot(x='Name', data=df)
plt.title("Frequency of Names")
plt.show()
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['Species'] = iris.target
df['Species'] = df['Species'].map({ 0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'})

sborn.kdeplot(data=df[df['Species'] == 'Virginica'], x='sepal length (cm)', fill=True, label='Virginica')
plt.legend()
plt.show()
```

```python
# customize seaborn plots
iris = sborn.load_dataset('iris')
sborn.scatterplot(x='sepal_length', y='sepal_width', data=iris)

plt.title("Sepal lengths vs Sepal width")
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')
plt.show()
```

```python
sborn.set_style("whitegrid")

sborn.boxplot(x='species', y='petal_length', data=sborn.load_dataset('iris'))
plt.title('Petal length distruibution by species')
```

```python
custom_colors = ['#FF5733', '#33FFBD', '#335BFF']
sborn.set_palette(custom_colors)

sborn.violinplot(x='species', y='petal_length', data=sborn.load_dataset('iris'))
plt.title('Custom Colored Petal Length Distribution')
plt.show()
```
