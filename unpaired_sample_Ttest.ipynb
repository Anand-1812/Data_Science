```python
import numpy as npy
from scipy import stats

A = npy.array([78,84,92,88,75,80,85,90,87,7978,84,92,88,75,80,85,90,87,79])
B = npy.array([82,88,75,90,78,85,88,77,92,8082,88,75,90,78,85,88,77,92,80])

alpha = 0.05
diff = len(A) + len(B) - 2

t_val, p_val = stats.ttest_ind(A, B)
t_critical = stats.t.ppf(1-alpha / 2, diff)

print(f"T-value: {t_val}")
print(f"P-value: {p_val}")
print(f"T-critical: {t_critical}")
```

```python
print('T-test Result:')
if npy.abs(t_val) > t_critical:
  print('Significant difference found.')
else:
  print('No significant difference.')

print('P-test Result:')
if p_val > alpha:
  print('Fail to reject H0. No strong evidence of difference.')
else:
  print('Reject H0. Significant difference found.')
```
