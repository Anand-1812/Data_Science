```python
import numpy as npy
import pandas as pd
from sklearn.svm import SVC
from sklearn.datasets import make_circles
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# making dataset

X, y = make_circles(n_samples=500, factor=0.5, noise=0.05, random_state=42)
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Creating and Training the Non-Linear SVM Model
svm = SVC(kernel='rbf', C=1, gamma=0.5)
svm.fit(X_train, y_train)
```

```python
# making predictions and evaluation model
y_pred = svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# visualizing decision boundary
import matplotlib.pyplot as plt

def plot_decision_boundary(X, y, model):
  x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
  y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

  xx, yy = npy.meshgrid(
    npy.arange(x_min, x_max, 0.01),
    npy.arange(y_min, y_max, 0.01)
  )

  Z = model.predict(npy.c_[xx.ravel(), yy.ravel()])
  Z = Z.reshape(xx.shape)

  plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
  plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.Paired)
  plt.title("Non-linear SVM with Polynomial Kernel")
  plt.show()

plot_decision_boundary(X, y, svm)
```

```python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=500, noise=0.1, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
poly_svm = SVC(kernel='poly', degree=3, C=1, coef0=1)
poly_svm.fit(X_train, y_train)
```

```python
y_pred = poly_svm.predict(X_test)
poly_accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {poly_accuracy:.2f}")
```
