```python
import numpy as npy
import scipy.stats as stat

cc = 1200 # control group clicks
ci = 10000 # control group impression

tc = 1500 # test group cliks
ti = 10000 # test group impression
```

```python
ctr_c = cc / ci
ctr_t = tc / ti

table = npy.array([[ci, ci-cc], [ti, ti-tc]])
chi_test, p, _, _ = stat.chi2_contingency(table)

print(f"Control CTR: {ctr_c:.2%}")
print(f"Test CTR: {ctr_t:.2%}")
print(f"Chi-Square Test p-value: {p:.5f}")

if p < 0.05:
  print("The difference is statistically significant.Implement the new recommendation system.")
else:
  print("No significant difference. Further testing needed.")
```
