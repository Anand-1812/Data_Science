```python
import sklearn
import pandas as pd

data_frame = pd.read_csv('customer_purchase_data.csv');
print(data_frame.head())
```

```python
print(data_frame.info())
print(data_frame.describe(include='all'))
```

```python
print(data_frame.isnull().sum())
data_frame = data_frame.dropna()

print(data_frame.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_feature = ['Gender', 'Age', 'AnnualIncome', 'ProductCategory']
for feature in categorical_feature:
  data_frame[feature] = le.fit_transform(data_frame[feature])

data_frame['Purchase'] = le.fit_transform(data_frame['PurchaseStatus'])
```

```python
from sklearn.feature_selection import chi2, SelectKBest

X = data_frame[categorical_feature]
Y = data_frame['Purchase']

selector = SelectKBest(score_func=chi2, k=2)
X_new = selector.fit_transform(X, Y)

feature_scores = selector.scores_
selected_feature = X.columns[selector.get_support()]

print(f"feature score: {feature_scores}")
print(f"Selected feature: {selected_feature}")
```
