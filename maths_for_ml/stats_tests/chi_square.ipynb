```python
import numpy as npy
import scipy.stats as stat
import matplotlib.pyplot as plt

observed = npy.array([20, 30, 40, 25, 10, 15])
expected = npy.array([25, 25, 32.5, 32.5,12.5, 12.5])

df = 2
alpha = 0.05

chi_stats = ((observed - expected) ** 2 / expected).sum()
critical_value = stat.chi2.ppf(1-alpha, df)

print(f"Calculated chi-square = {chi_stats:.3f}")
print(f"Critical value (chi2, df={df}, alpha={alpha}) = {critical_value:.3f}")
```

```python
x = npy.linspace(0, 10, 1000)
y = stat.chi2.pdf(x, df)

plt.plot(x, y, "Chi-Square (df=2)")
plt.fill_between(x, y, where=(x > critical_value), color="red", alpha=0.5, label='Critical region')
plt.axvline(chi_stats, color='blue', linestyle='dashed', label='Calculated Chi-Square')
plt.axvline(critical_value, color='green', linestyle='dashed', label='Critical Value')

plt.title("Ch-Square and Critical regin")
plt.xlabel("Chi square value")
plt.ylabel("Probability Density Function")
plt.legend()
plt.show()
```
