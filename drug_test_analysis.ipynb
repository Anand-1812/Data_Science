```python
import math
import numpy as np
import statistics as st

before_test_data = [120, 122, 118, 130, 125, 128, 115, 121, 123, 119]
after_test_data = [115, 120, 112, 128, 122, 125, 110, 117, 119, 114]

significance_level = 0.05
n = len(before_test_data)
```

```python
differences = [after - before for after, before in zip(after_test_data, before_test_data)]

mean_diff = np.mean(differences)
sd_diff = st.stdev(differences)

t_value = mean_diff / (sd_diff / math.sqrt(n))
```
