```python
import numpy as npy
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generating random data set

npy.random.seed(42)
X = npy.random.rand(50, 1)*100
y = 3.5*X + npy.random.randn(50, 1)*20
```

```python
# Create and Train linear model

linear_model = LinearRegression()
linear_model.fit(X, y)

# predicting y-values

Y_pred = linear_model.predict(X)
```

```python
# Visualize the regression line

plt.figure(figsize=(10, 8))
plt.scatter(X, y, color='green', label='Data Point')
plt.plot(X, Y_pred, color='red', linewidth=2.5, label='Regression line')

plt.title("Linear Regression on random dataset")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Slope: ", linear_model.coef_[0][0])
print("Intercept: ", linear_model.intercept_[0])
```
