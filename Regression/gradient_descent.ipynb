```python
import numpy as npy
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=5, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
plt.scatter(X, y, color='green')
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)

LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None)
print(lr.coef_, lr.intercept_)

plt.scatter(X, y, color='green')
plt.plot(X, lr.predict(X), color='red')
```

```python
# Now apply gradient descent assuming slope is m = 78.3, and intercept b = 0

y_pred = ((78.3*X) + 0).reshape(5)

plt.scatter(X, y, color='green', label='OLS')
plt.plot(X, lr.predict(X), color='red', label='b = 0')
plt.plot(X, y_pred, color='blue')
plt.legend()
plt.show()
```

```python
m = 78.3
b = 0
l_rate = 0.01

loss_slope = -2 * npy.sum(y-m*X.ravel()-b)

step_size = loss_slope * l_rate
b = b - step_size

y_pred1 = (m*X + b).reshape(5)

plt.scatter(X, y, color='green', label='OLS')
plt.plot(X, lr.predict(X), color='red', label='b updated')
plt.plot(X,y_pred1, color='blue', label='b = {}'.format(b))
plt.plot(X,y_pred, color='orange', label='b = 0')
plt.legend()
plt.show()
```
