```python
import numpy as npy
import scipy.stats as stat

A = npy.array([4, 4, 7, 16, 20, 11, 13, 9, 11, 15])
B = npy.array([15, 16, 14, 14, 22, 22, 23, 18, 18, 19])

t_val, p_val = stat.ttest_rel(A, B)
alpha = 0.05
dof = len(A) - 1

t_critical = stat.t.ppf(1-alpha/2, dof)
print("T-value:", t_val)
print("P-Value:", p_val)
print("Critical t-value:", t_critical)
```

```python
print('T-test:')
if npy.abs(t_val) > t_critical:
  print('Significant difference found.')
else:
  print('No significant difference.')

print('P-test:')
if p_val > alpha:
  print('Reject H0')
else:
  print('Fail to reject H0')
```
